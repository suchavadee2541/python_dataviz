# Barchart (กราฟแท่ง)

## สร้างกราฟแท่งเปรียบเทียบรายได้ในปี 2541 ของจังหวัดกรุงเทพและปริมณฑล

###### x คือตำแหน่งในแนวนอน  

```python
plt.bar([1,2,3,4],greaterbankok.iloc[:,2])  #แท่งที่1อยู่1 แท่ง2อยู่2  iloc เอาทุกแถว:คอลัมที่2
plt.title(u'กราฟแท่งเปรียบเทียบรายได้ในปี 2541 ของจังหวัดกรุงเทพและปริมณฑล',fontname = 'Tahoma',fontsize=18) #ใส่ชื่อตาราง
```

```python
plt.bar([1,2,3,4],greaterbankok.iloc[:,2])  #แท่งที่1อยู่1 แท่ง2อยู่2  iloc เอาทุกแถว:คอลัมที่2
plt.title('Barchart compareing income of provinces in Greater Bankok in 2541') #ใส่ชื่อตาราง
```

การกำหนดชื่อแกน คือ xlebel

```python
plt.bar([1,2,3,4],greaterbankok.iloc[:,2])  #แท่งที่1อยู่1 แท่ง2อยู่2  iloc เอาทุกแถว:คอลัมที่2
plt.title('Barchart compareing income of provinces in Greater Bankok in 2541') 
plt.xlabel('Provinces') #กำหนดชื่อแกนX
plt.ylabel('Income Bath/Month')#กำหนดชื่อแกนY
```

## add tick label  การใส่ชื่อ

```python
plt.bar([1,2,3,4],greaterbankok.iloc[:,2],tick_label=greaterbankok.iloc[:,-1])  #แท่งที่1อยู่1 แท่ง2อยู่2  iloc เอาทุกแถว:คอลัมที่2
plt.title('Barchart compareing income of provinces in Greater Bankok in 2541') 
plt.xlabel('Provinces') #กำหนดชื่อแกนX
plt.ylabel('Income Bath/Month/Household')#กำหนดชื่อแกนY
```

## Compare different years การเปรียบเทียบกราฟ

```python
plt.bar([1-(0.35/2),2-(0.35/2),3-(0.35/2),4-(0.35/2)],greaterbankok.iloc[:,-2],width=0.35,tick_label=greaterbankok.iloc[:,-1],color='g',label = '2541',align='edge') # align คือการขยับlabel
plt.bar([1+(0.35/2),2+(0.35/2),3+(0.35/2),4+(0.35/2)],greaterbankok.iloc[:,2],width=0.35,tick_label=greaterbankok.iloc[:,-1],color='b',label = '2558',align='edge')
plt.title('Barchart compareing income of provinces in Greater Bankok in 2541,2558') 
plt.xlabel('Provinces') #กำหนดชื่อแกนX
plt.legend() # เลือกตำแหน่งที่ต้องการ
plt.ylabel('Income Bath/Month/Household')#กำหนดชื่อแกนY
```
