## Visualize arrey

```python
from matplotlib import pyplot as plt
import matplotlib
```

```python
plt.imshow(setosa_np,interpolation='nearest')
plt.show
```

สีสว่างค่ามาก สีข้มค่าน้อย

```python
matplotlib.rcParams['figure.figsize'] = [12,12]
plt.imshow(setosa_np,interpolation='nearest')
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = [12,12]
plt.imshow(vericolor_np,interpolation='nearest')
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = [12,12]
plt.imshow(veriginica_np,interpolation='nearest')
plt.show()
```

subplot คือ plotหลายๆรูปไว้ในรูปเดียว
plt.subplot(แถว1,3หลัก,ตำแหน่ง1)

```python
matplotlib.rcParams['figure.figsize'] = [12,12]
plt.subplot(1,3,1)
plt.imshow(setosa_np,interpolation='nearest')
plt.subplot(1,3,2)
plt.imshow(vericolor_np,interpolation='nearest')
plt.subplot(1,3,3)
plt.imshow(veriginica_np,interpolation='nearest')
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = [12,12]
plt.subplot(3,1,1)
plt.imshow(setosa_np,interpolation='nearest')
plt.subplot(3,1,2)
plt.imshow(vericolor_np,interpolation='nearest')
plt.subplot(3,1,3)
plt.imshow(veriginica_np,interpolation='nearest')
plt.show()
```
