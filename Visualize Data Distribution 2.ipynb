# Box Plot

```python
import pandas as pd
import matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
irisdata = pd.read_csv('C:\\Users\\computer\\Desktop\\Work\\T.Ton\\Phython\\python_dataviz\\iris.data',header=None)
irisdata
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()
virginica_np = irisdata.iloc[100:,:-1].to_numpy()
```

Q1=Quartile 1 คือ แบ่งออกเป็น 4ส่วน โดย Q คือตัวที่25  ค่า median คือ Q2 ถ้าQ2เข้าใกล้Q1ค่าที่มากกว่าQ2สูงมาก 
IQR คือ เอาQ3-Q1
whisker คือ ตัวที่ช่วยระบุ dataเราเลยค่าที่กำหนดตัวนั้นจะเป็น out lier
บอกการกระจายของข้อมูล

```python
matplotlib.rcParams[ 'figure.figsize'] = [15, 5 ]   #ดูการกระจาย
plt.subplot(1, 4, 1)   #มี1 แถว / 4 หลัก / อยู่ตำแหน่งที่ 1
plt.boxplot(setosa_np[:, 0]) 
plt.subplot(1, 4, 2)  
plt.boxplot(setosa_np[:, 1]) 
plt.subplot(1, 4, 3)
plt.boxplot(setosa_np[:, 2]) 
plt.subplot(1, 4, 4)
plt.boxplot(setosa_np[:, 3]) 
plt.show()
```

```python
import pandas as pd 
Debt ='Average Debt.xls'
Debt_df=pd.read_excel(Debt)
Debt_df=pd.read_excel(Debt,skiprows=3)
```

```python
Debt_df
```

#### เลือก column ที่เราสนใจ

```python
Debt_df[['Region and province',2549]]
```

```python
Debt_df_debt = Debt_df[['Region and province',2549]]
```

#### ทำการลบแถวที่ไม่เกี่ยวข้องออกไป

```python
Debt_df_debt.drop([0,1,2,3,8,31,49,70,85,86,87,88],axis = 0,inplace = True)
```

```python
Debt_df_debt.head()
```

### ทำการimport folium เพื่อใช้สำหรับการทำแผนที่

```python
import folium
import pandas as pd
```

```python
data_to_plot = Debt_df_debt
```

```python
data_to_plot.head()
```

### ทำการDownloadข้อมูลเชิงภูมิศาสตร์ของประเทศไทยจากเว็บไซต์

```python
country_geo = 'thai.json'
```

```python
new_col = list()
for d in data_to_plot['Region and province']:
    new_col.append(d.lstrip())
data_to_plot['Region and province'] = new_col
```

### ทำการกำหนดพิกัด ลองติจูดและละติจูดของประเทศไทย

```python
map =  folium.Map(location = [10,100],zoom_start = 5)
```

```python
map
```

### ทำการลงสีของข้อมูล

```python
map.choropleth(geo_data = country_geo, data=data_to_plot,
              columns=['Region and province',2549],
               key_on='feature.properties.name',
              fill_color='YlGnBu',fill_opacity=0.7,line_opacity=0.2, #ลงสีของข้อมูล
              legend_name='Average Debt')
```

```python
map.save('plot_data.html') #การ seve ไฟล์เป็นแบบ HTML อินเตอร์เน็ต
```
