## catter plot เป็นกราฟแบบกระจัดกระจายเป็นจุดๆ ส่วนใหญ่เป็น 2 มิติ 

```python
irisdata = pd.read_csv('C:\\Users\\computer\\Desktop\\Work\\T.Ton\\Phython\\python_dataviz\\iris.data',header=None)
irisdata
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
setosa_np[:5,:]
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1])  #ตัวแรกในวงเล็บบอกแถว > เอาทุกแถว เอาหลักที่0 
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=100) # S บอกขนาดของจุด  
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=setosa_np[:,2]) # s เป็นตัวที่3 เลยใช้ 2
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2],c='g')  #c,m,y,k สีของจุด
```

```python
setosa_np
```

```python
print(f'max={max(setosa_np[:,3])},min ={min(setosa_np[:,3])}') #หาค่าmin and max
```

```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i<0.3:
        colors.append('g')
    elif i<0.45:
        colors.append('y')
    else:
        colors.append('r')
colors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=colors)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=setosa_np[:,3])#ไม่ใส่ค่าcolors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=setosa_np[:,3],cmap = 'YlOrBr')  
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=np.array([[1,0.5,0.8]]))  #ใส่สี โดยเลข1คือใส่สีนั้นเต็มๆ
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=setosa_np[:,3],cmap = 'YlOrBr',marker='p')
```
